```python
import numpy as np
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from wrangler import Wrangler

from sktrajectory import SKTrajectory
from trajectory import Trajectory
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

from PIL import Image, ImageOps
from matplotlib.animation import FuncAnimation, writers

import xgboost as xgb

import matplotlib as mpl 
mpl.rcParams['animation.ffmpeg_path'] = r'C:\ffmpeg\bin\ffmpeg.exe'
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
np.set_printoptions(suppress=True)
```

```python
pdf = Wrangler.load_pickle('data/pdf.pkl')
pdf = pdf[pdf['cluster'] == 5]
nndf = Wrangler.load_pickle('data/nndf.pkl')
nndf = nndf.loc[nndf['cluster'] == 5]
l_df = pd.read_csv('bsc-3m/signals_dense.csv')
l_xy = Wrangler.load_pickle('bsc-3m/signal_lines_true.pickle')
```

```python
# clf = Wrangler.load_pickle('models/model.pkl')
cols = ['x', 'y', 'd_t-1', 'd_t-2', 'd_t-3', 'd_light', 'l0', 'l1',
                'l2', 'l3', 'dir_0', 'dir_1', 'dir_2', 'target']
df_train, df_val = train_test_split(nndf[cols], test_size=0.2, random_state=1)
df_train, df_test = train_test_split(df_train, test_size=0.2, random_state=1)

x_train, y_train = df_train[cols[:-1]].to_numpy(), df_train[cols[-1:]].to_numpy().reshape(-1)
x_val, y_val = df_val[cols[:-1]].to_numpy(), df_val[cols[-1:]].to_numpy().reshape(-1)
x_test, y_test = df_test[cols[:-1]].to_numpy(), df_test[cols[-1:]].to_numpy().reshape(-1)

# print('training model...')
# clf = MLPRegressor().fit(x_train, y_train)
# score = clf.score(x_val, y_val)
# print('validation score:\t', score)
```

```python
reg = xgb.XGBModel()
reg.fit(x_train, y_train, eval_set=[(x_val, y_val)], eval_metric='rmse', verbose=True)
```

```python
pred = reg.predict(x_test)
r2_score(y_test,pred)
```

```python
Trajectory.full_sim_data = pd.DataFrame(columns=['x', 'y', 'd_t-1', 'd_t-2', 'd_t-3', 'd_light', 'l0', 'l1',
                                          'l2', 'l3', 'dir_0', 'dir_1', 'dir_2', 'frame', 'class'])

fig, ax = plt.subplots(figsize=(20,20))
ax.set_xlim(0,1280)
ax.set_ylim(0,720)
img = Image.open("intersection2.png")
img = ImageOps.flip(img)

im = ax.imshow(img, origin='lower')
ln = [ax.plot([],[])[0] for _ in range(l_df.shape[-1]-1)]
txt = ax.text(20,20, '', fontsize=35, color='w')
sc = ax.scatter([], [], s=200)
patches = ln + [sc] + [txt]

num_t = 10
ts = [SKTrajectory(pdf.iloc[i], l_df, l_xy, reg) for i in range(num_t)]
for i in range(num_t):
    ts[i].init_sim(0)
ts_ln = [ax.plot([],[])[0] for _ in range(num_t)]

def init():
    for i in range(len(ln)):
        ln[i].set_data(l_xy[i]['x'], l_xy[i]['y'])
    
    return patches

def update(frame):
    row = l_df.loc[frame]
    for i in range(l_df.shape[1]-1):
        ln[i].set_color(['red', 'orange', 'yellow', 'green'][row[str(i)]])
    
    txt.set(text=str(frame))
    
    for i in range(num_t):
        ts[i].step(frame)
        
    sc.set_offsets(ts[0].full_sim_data[['x','y']].to_numpy())
                   
    return patches

ani = FuncAnimation(fig, update, frames=np.arange(10,2000,5),    # np.arange(0, 12000, 10), # interval=30,
                    init_func=init)#, blit=True)

Writer = writers['ffmpeg']
writer = Writer(fps=6) # , bitrate=1800)
ani.save('predict3.mp4', writer)
```
